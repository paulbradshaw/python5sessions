# Reshaping data with two header rows

```python
import pandas as pd
#read in the data - the header is row 2
df = pd.read_excel("epc_lad_counts_cfy_10-21.xlsx", header=1)
df.head()
```

```python
#read in just row 1 and nothing else
years = pd.read_excel("epc_lad_counts_cfy_10-21.xlsx", skiprows=0, nrows=0, dtype=['str'])
#convert to a list, and to strings
years = list(map(str, list(years.columns)))
#show
print(years)
#replace the 00 timestamp in all items
years = [yr.replace(' 00:00:00', '') for yr in years]
print(years)
```

## Encode that in a loop

Now we've tested the concept, we can encode it in a loop.

We just need to loop through those indices, and assign to the column at that position the date in the position 7 earlier in the first row from the original spreadsheet.

```python
#loop through the list of indices
for i in codecols:
  print(i)
  #use that to access the column at that index, and fill it with the date 7 positions previous in the first row we grabbed
  df.iloc[:,i] = years[i-7]
  #check
  df.iloc[:,i]
```

## Alternative approach - embed years in column names

We might notice that the columns are named with a period and a number to avoid duplicates - but this also allows us to connect them to a year. For example:

* A.1 and B.1 etc are 2011-12
* A.2 and B.2 etc are 2012-13

So we can replace the period (.) with 201 and get 

* A2011
* A2012

Or something along those lines.

```python
#start all over again
import pandas as pd
#read in the data - the header is row 2
df = pd.read_excel("epc_lad_counts_cfy_10-21.xlsx", header=1)
df.head()
```

```python
df.columns
```

```python
[i.replace('.',' 201') for i in df.columns]
```

This doesn't quite work for the last one. We could fix that with a second block of code.

```python
#replace the columns with the cleaned versions
df.columns = [i.replace('.',' 201') for i in df.columns]
#then use to replace the last year error
df.columns = [i.replace('20110','2020') for i in df.columns]
df.columns
```

And delete the duplicate columns

```python
#create a True/False list of which columns contain 'lad21cd'
lacols = ['lad21cd' in i for i in df.columns]
#fetch just those columns
lacolnames = df.columns[lacols]
#check
print(lacolnames)
#remove the first, as we'll want to keep that
lacolnames = lacolnames[1:]
```

```python
#drop those columns
df.drop(axis=0, columns=lacolnames)
```

```python
#now save the results in the same dataframe
df = df.drop(axis=0, columns=lacolnames)
#check it's worked
df.columns
```

## Reshape from wide to long

Now we can reshape the data.

First we use `.melt()` to convert from wide to long, in a very basic way.

```python
#https://pandas.pydata.org/docs/reference/api/pandas.melt.html
longdf = pd.melt(df,id_vars='lad21cd', value_name='value')
print(longdf)
```

This results in a very long dataframe - but allows us to split out the years en masse. 

```python
print(longdf['variable'].value_counts())
```

```python
#split out the codes
codes = [i.split(' ')[0] for i in longdf['variable']]
codes[-10:]
```

```python
#add back into dataframe
longdf['code'] = codes
longdf
```

The year is trickier because there's no year on the first set of codes, so a `.split()` function would return an error on those if you tried to access the second item after a split (the year)

```python
[i.split(' ')[1] for i in longdf['variable']]
```

Instead, then we could grab the last four characters. This will tolerate a situation where there's only one character.

```python
years = [i[-4:] for i in longdf['variable']]
print(years[:10])
print(years[-10:])
```

But this still leaves us with those single characters.

We can use a `for` loop and an `if` test to create a new list which either keeps the year or, if it's a single character, replaces it with '2010'.

```python
#create an empty list to fill
fullyrs = []
#loop through our list of years
for i in years:
  #if it's not a year, but a single character
  if len(i) == 1:
    #replace with the '2010' - this time a number
    fullyr = 2010
  #if it's a year
  else:
    #just keep it, but convert to a number (integer)
    fullyr = int(i)
  #then append whichever it was to the previously empty list
  fullyrs.append(fullyr)
print(fullyrs[:10])
print(fullyrs[-10:])
```

```python
longdf['year'] = fullyrs
longdf.head()
```

```python
longdf.to_csv("longdf.csv")
```
